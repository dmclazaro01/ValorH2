```python
import spacy
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import json

# Cargar modelo de lenguaje de Spacy
nlp = spacy.load("en_core_web_md")


# Descripción del IOC
ioc_description = "URL that delivers a malware payload"



# Definir técnicas y sus descripciones
tactics_descriptions = {}


with open("alt/mittre_all.json") as f:
    tactics = json.load(f)
    print(tactics.keys())
    for tactic,data in tactics.items():
        if tactic=="count":
            print
        else:
            
            tactics_descriptions[tactic] = tactics[tactic].get("description","")
        
        
# Tokenizar y procesar la descripción del IOC
doc_ioc = nlp(ioc_description)

# Procesar las descripciones de las técnicas
docs_tactics = [nlp(tactics_description) for tactics_description in tactics_descriptions.values()]

# Calcular la similitud de coseno entre la descripción del IOC y cada técnica
vectorizer = CountVectorizer().fit_transform([doc_ioc.text] + [doc_tactic.text for doc_tactic in docs_tactics])
similarity_matrix = cosine_similarity(vectorizer)

# Encontrar la técnica más cercana
best_match_index = similarity_matrix[0][1:].argmax()
best_match = list(tactics_descriptions.keys())[best_match_index]

# Imprimir la mejor coincidencia
print(f"La técnica más cercana es: {best_match} con una similitud de {similarity_matrix[0][best_match_index + 1]}")
```

```python
import requests
import json
from bs4 import BeautifulSoup
import re


URL = "https://attack.mitre.org/"
URL_TECHNIQUES = URL + "techniques/"
URL_TACTICS = URL + "tactics/"

def get_desc(id): #Obtiene la descripcion de la tactica
    url = URL_TACTICS + id + "/"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    description = soup.find("div",{"class":"description-body"})
    description_text = description.find_all("p")
    return description_text[0].text

def get_desc_tech(id): #Obtiene la descripcion de la tecnica
    url = URL_TECHNIQUES + id + "/"
    response = requests.get(url)
    
    soup = BeautifulSoup(response.text, "html.parser")
    description = soup.find("div",{"class":"description-body"})
    description_text = description.find_all("p")
    return description_text[0].text
    
def get_all(return_data):
    pattern = r'\d+'
    tactics_count = []
    tactics_name = []
    tactics_desc = []
    tactics_id = []
    techniques_name = []
    techniques_desc = []
    mittre = {}
    subtechniques = {}
    response = requests.get(URL)
    soup = BeautifulSoup(response.text, "html.parser")
    matrix = soup.find("div", {"class": "matrix-container p-3"})
    tactics_html = matrix.find_all("td", {"class": "tactic name"})
    for i in tactics_html:
        name = i.find("a").text
        tactid = i.find("a")["title"]
        tactics_name.append(name)
        tactics_id.append(tactid)
        tactics_desc.append(get_desc(tactid))
    tactics_count_html = matrix.find_all("td", {"class": "tactic count"})
    for i in tactics_count_html:
        number = re.findall(pattern, i.text)
        tactics_count.append(int(number[0]))
    techniques_html = matrix.find_all("table", {"class": "techniques-table"})
    for tactic in tactics_name:
        mittre[tactic] = {"tactid": tactics_id[tactics_name.index(tactic)], "description": tactics_desc[tactics_name.index(tactic)]}
    index = 0
    for techniques_colum in techniques_html:
        techniques_rows = techniques_colum.find_all("tr", {"class": "technique-row"})
        for techniques_row in techniques_rows:
                supertechnique = techniques_row.find("table",{"class": "supertechnique"})
                if (supertechnique == None):
                    technique_cell = techniques_row.find("div", {"class": "technique-cell"})
                    technique_cell_a = technique_cell.find("a")
                    technique_name = technique_cell_a.text
                    techniques_name.append(technique_name)
                    tid = technique_cell_a["title"]
                    data = {"tid": tid, "description": get_desc_tech(tid)}
                    

                    mittre[tactics_name[index]][technique_name] = data
                else:
                    technique_cell = supertechnique.find("div")
                    technique_cell_a = technique_cell.find("a")
                    technique_name = technique_cell_a.text
                    technique_name = technique_name.split("\xa0")[0]
                    techniques_name.append(technique_name)
                    
                    tid = technique_cell_a["title"]
                    techniques_desc.append(get_desc_tech(tid))
                    subtechniques_html = techniques_row.find_all("div", {"class": "subtechnique"})
                    subtechniques = {}
                    for subtechnique in subtechniques_html:
                        subtechnique_a = subtechnique.find("a")
                        subtechnique_name = subtechnique_a.text
                        s_tid = subtechnique_a["title"]
                        subtechniques[subtechnique_name] = {"stid": s_tid}
                    data = {"tid": tid, "description": get_desc_tech(tid), "subtechniques": subtechniques}
                    mittre[tactics_name[index]][technique_name] = data
        mittre[tactics_name[index]]["count"] = tactics_count[index]
        index += 1
    mittre["count"] = len(tactics_count)
    json_string = json.dumps(mittre,indent=4)
    print(json_string)
    with open("alt/mittre_all.json", "a") as outfile:
        
        outfile.write(json_string)
        outfile.write(",\n")
    #if (return_data == 0):
    #    return mittre
    #elif (return_data == 1):
    #    return tactics_name,tactics_id
    # elif (return_data == 2):
    #    return techniques_name
    
print("Obteniendo datos de MITRE ATT&CK...")    
get_all(0)
```
